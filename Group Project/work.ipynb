```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
movie_data = pd.read_csv('Movies.csv')
```

```python
print("The number of movies for each genre:\n")
print(movie_data['Genre'].value_counts())
```

```python
print("The percentage of movies for each genre:\n")
print(movie_data['Genre'].value_counts() * 100 / len(movie_data))
```

```python
print("Graph of genres:\n")
movie_data['Genre'].value_counts().plot(kind='bar', title="Genre")
```
